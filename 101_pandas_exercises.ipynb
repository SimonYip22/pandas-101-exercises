## 7. How to get the items not common to both series A and series B?

Difficulty Level: L2

Get all items of ser1 and ser2 not common to both.

Desired Output:

0    1
1    2
2    3
5    6
6    7
7    8
dtype: int64

```python
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

# Solution to find symmetric difference
ser_u = pd.Series(np.union1d(ser1, ser2)) # find the union
ser_i = pd.Series(np.intersect1d(ser1, ser2)) # find the intersect
# from the union, remove the intersect to get the difference
print(ser_u[~ser_u.isin(ser_i)])

# or

only_in_ser1 = ser1[~ser1.isin(ser2)] # find the difference in ser1
only_in_ser2 = ser2[~ser2.isin(ser1)] # find the difference in ser2

# concatenate the two series to get the symmetric difference
result = pd.concat([only_in_ser1, only_in_ser2])
```

## 8. How to get the minimum, 25th percentile, median, 75th, and max of a numeric series?

Difficuty Level: L2 Compute the minimum, 25th percentile, median, 75th, and maximum of ser.

Desired Output:

array([ -1.39,   6.49,  10.26,  13.07,  25.81])
# (exact values vary — output shows the 5 percentiles)

```python
ser = pd.Series(np.random.normal(10, 5, 25))

# using numpy percentile method
print(np.percentile(ser, q=[0, 25, 50, 75, 100]).round(2))

# or

# using pandas quantile method
result = ser.quantile([0, 0.25, 0.5, 0.75, 1]).round(2).to_numpy()

result = ser.describe()[["min", "25%", "50%", "75%", "max"]].round(2)

print(result)
```

## 10. How to keep only top 2 most frequent values as it is and replace everything else as ‘Other’?

Desired Output:

Top 2 Freq: 4    5

...

0     Other
1     Other
2         3
...

```python
np.random.seed(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

top_2 = ser.value_counts().index[:2] # get the top 2 frequent values

result = ser.where(ser.isin(top_2), "Other")

print("Top 2 Freq:", top_2)

print(result)
```

## 11. How to bin a numeric series to 10 groups of equal size?

Difficulty Level: L2

Bin the series ser into 10 equal deciles and replace the values with the bin name.

Desired Output:

...
dtype: float64

0    7th

...
dtype: category
Categories (10, object): [1st < 2nd < 3rd < ... < 9th < 10th]

```python
ser = pd.Series(np.random.random(20))

# Solution
# quartile binning is done using qcut where the bins have equal counts of data points.
print(pd.qcut(ser, q=10, 
labels=['1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th', '10th']).head())
```

## 17. How to compute the mean squared error on a truth and predicted series?

Difficulty Level: L2

Compute the mean squared error of truth and pred series.

```python
truth = pd.Series(range(10))
pred = pd.Series(range(10)) + np.random.random(10)

# MSE used in ML regression models.
# Common regression metric to evaluate the performance of a model. 
# It is the average of the squared differences between the predicted and actual values.
# This metric is used to measure how close a regression line is to a set of points.
# lower MSE = better model performance, higher MSE = worse model performance.


difference = truth - pred 
squared_difference = difference ** 2
mse = squared_difference.mean()

print(mse)
```

## 22. How to get the day of month, week number, day of year and day of week from a series of date strings?

Difficulty Level: L2 Get the day of month, week number, day of year and day of week from ser.

```python
ser = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])

# Solution
ser_ts = pd.to_datetime(ser, format="mixed")

# day of month
print("Date: ", ser_ts.dt.day.tolist())

# week number
print("Week number: ", ser_ts.dt.isocalendar().week.tolist())

# day of year
print("Day num of year: ", ser_ts.dt.dayofyear.tolist())

# day of week
print("Day of week: ", ser_ts.dt.day_name().tolist())
```

## 26. How to get the mean of a series grouped by another series?

Difficiulty Level: L2 Compute the mean of weights of each fruit.

```python
fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weights = pd.Series(np.linspace(1, 10, 10))

print(weights.groupby(fruit).mean())

df = pd.DataFrame({
    "fruit": fruit,
    "weight": weights
})

result = df.groupby("fruit")["weight"].mean()

print(result)
```
